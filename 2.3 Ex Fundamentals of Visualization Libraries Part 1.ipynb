### Basic plotting with pandas

```python
# setting date as the index
df_temp = df.set_index('date')
```

#### Code Explanation

- In this step, I created the daily temperature using matplotlib that used object hierarchy. I first defined figure size and then provided the title. I then created chart using plt.plot code and then showed (and hide the description) the plot using plt.show()

# Trips per Day

```python
# Groupby the count of the trips 

df_group = pd.DataFrame(df.groupby(['date'])['ride_id'].count()).reset_index()
```

```python
df_group.head()
```

```python
df_group.rename(columns = {'ride_id':'bike_rides_daily'}, inplace = True)
df = df.merge(df_group, on = "date", how = 'outer', indicator = True)
print(df['_merge'].value_counts(dropna = False))
print("Shape of January 1st is", df[df['date'] == '2022-01-01'].shape) # Check 
print("Shape of January 2nd is", df[df['date'] == '2022-01-02'].shape) # Second check 

df_temp = df.set_index('date')

print(df_temp.columns)
```

```python
df_temp.head()
```

```python
plt.figure(figsize=(10,5))
plt.title('Temperature per trips')
plt.plot(df_temp['avgTemp'])
plt.plot(df_temp['avgTemp'], color = 'red')
plt.show()
```

#### Code Explanation

- In this step, we added the color of the chart to make it red by assigning the color in the code

```python
# Create figure and axis objects with subplots()

fig,ax = plt.subplots(figsize=(10, 5))

# Make a plot of the bike rides
ax.plot(df_temp['bike_rides_daily'],color = "red")
```

#### Code Explanation

- In this step, we changed the chart from Temperature to Bike rides

```python
# Create figure and axis objects with subplots()

fig,ax = plt.subplots(figsize=(10, 5))

# Make a plot of the bike rides
ax.plot(df_temp['bike_rides_daily'],color = "red")
ax.plot(df_temp['avgTemp'], color = 'blue')
```

#### Code Explanation

- In this step, we created a combo chart of both bike rides (red) and temperature (blue). The temperature came as straight line because of the scale. The maximum value for average temperature was 32 but the chart's y-axis ranged from 0 to 140000 

```python
# Create figure and axis objects with subplots()

fig,ax = plt.subplots(figsize=(10, 5))

# Make a plot of the bike rides
ax.plot(df_temp['bike_rides_daily'],color = "red")

# Here is where we create the dual axis - initialize a twin object for two different y-axis on the sample plot
ax2 = ax.twinx()

# Make a plot with the temperatues on the y-axis using second axis object
ax2.plot(df_temp["avgTemp"], color = "blue")
plt.show()
```

#### Code Explanation

- In this step, I fixed the above problem by specifying secondary axis for Temperature. Now the axis for blue line is to the right while the axis for the red line is in the left

```python
# Create figure and axis objects with subplots() and details

fig,ax = plt.subplots(figsize=(10, 5))

# Make a plot of the bike rides
ax.plot(df_temp['bike_rides_daily'],color = "navy")
ax.set_xlabel("Year 2022", fontsize = 14) # set x-axis label
ax.set_ylabel("Bike rides daily",color = "navy", fontsize = 14) # set y-axis label

# Here is where we create the dual axis - initialize a twin object for two different y-axis on the sample plot
ax2 = ax.twinx()

# Make a plot with the temperatues on the y-axis using second axis object
ax2.plot(df_temp["avgTemp"], color = "red")
ax2.set_ylabel("Average temperatures", color = "red",fontsize=14)
plt.title('Temperature and trips in 2022', fontsize = 18)
plt.show()
```

#### Code Explanation

- In this step, cleaned up the chart by adding Titles, and axis titles to make it clear which axis represent what. Also we changed the color as the temperatures are usually shown in red and bike rides in blue. The axis titles were put same colors to indicate which line represents what color

```python
# Create figure and axis objects with subplots() and details

fig,ax = plt.subplots(2, figsize =(8, 6))

# Make a plot of the bike rides
ax[0].plot(df_temp['bike_rides_daily'],color = "navy")
ax[0].set_ylabel("Bike rides daily",color = "navy", fontsize = 14) # set y-axis label

# Make a plot with the temperatues on the y-axis using second axis object
ax[1].plot(df_temp["avgTemp"], color = "red")
ax[1].set_ylabel("Average temperatures", color = "red",fontsize=14)
plt.suptitle('Temperature and trips in 2022', fontsize = 18)

plt.show()
```

#### Code Explanation

- In this step, we broke down the chart into two stacked line chart. I used subplot method in matplot lib to create two frames and then specified what data to use in first and second frame by using ax[0] (first) and ax[1] (second) .
