```python
#Import Libraries
import pandas as pd
from matplotlib import pyplot as plt
import numpy as np
from datetime import datetime as dt
import seaborn as sns
```

```python
#Import the files
df = pd.read_csv('NY_CityBike_Temp_2022.csv', index_col = 0)
```

```python
#Check Datatypes
df.dtypes
```

```python
#Set Theme for the Seaborn
sns.set_theme(style="ticks")
```

```python
#Check data
df.head()
```

## Creating Bar Plot using Seaborn

```python
#Assign each row as value 1 to make it easier to group
df['value'] = 1
```

```python
#Group the data to get count of the rides for each stations
df_t = df.groupby('start_station_name')['value'].count().reset_index()
```

```python
df_t
```

```python
#Filter top 20 rides based on count
top20 = df_t.nlargest(20, 'value')
```

```python
#Create sns barplot based on the default theme
sns.barplot(data = top20, x = 'value', y = 'start_station_name')
plt.title("Most frequent start stations")
plt.show()
```

```python
#change the palette to crest and do it in reverse order to show darker color for higher value
sns.barplot(data = top20, x = 'value', y = 'start_station_name', palette=("crest_r"))
plt.title("Most frequent start stations")
plt.show()
```

## Creating Line Plot using Seaborn

```python
#Convert date to datetime
df['date'] = pd.to_datetime(df['date'], format = '%Y-%m-%d')
```

```python
# Groupby the count of the trips 

df_group = pd.DataFrame(df.groupby(['date'])['ride_id'].count()).reset_index()
```

```python
#Check the data
df_group.head()
```

```python
#Create a new dataframe to get daily bike rides based on temperature
df_group.rename(columns = {'ride_id':'bike_rides_daily'}, inplace = True)
df = df.merge(df_group, on = "date", how = 'outer', indicator = 'merge_flag')
print(df['merge_flag'].value_counts(dropna = False))
print("Shape of January 1st is", df[df['date'] == '2022-01-01'].shape) # Check 
print("Shape of January 2nd is", df[df['date'] == '2022-01-02'].shape) # Second check 

df_temp = df.set_index('date')

print(df_temp.columns)
```

```python
#Sort dataframe by date
df_temp.sort_index(inplace=True)
```

```python
#Create the linechart based on sns library
fig,ax = plt.subplots(figsize=(10, 5))
sns.lineplot(data = df_temp['bike_rides_daily'], color = "b")
ax.set_xlabel("Year 2022", fontsize = 14) # set x-axis label
ax.set_ylabel("Bike rides daily",color = "blue", fontsize = 14) # set y-axis label

ax2 = ax.twinx()
sns.lineplot(data = df_temp['avgTemp'], color="r", ax=ax2)
ax2.set_ylabel("Average temperatures", color = "red",fontsize=14)
plt.title('Temperature and trips in 2022', fontsize = 18)
plt.show()
```

## Creating Box plots for member_casual

```python
df.dtypes
```

```python
#check the values on the column member_casual
df['member_casual'].value_counts()
```

```python
df.columns
```

```python
#Create boxplot using member_casual and average temperature
plt.figure(figsize=(9,5))
sns.boxplot(x = 'member_casual', y = 'avgTemp', data = df, palette = {'member': 'royalblue', 'casual': 'pink'})
plt.show()
```

Since we don't have other numerical variables, I used the average temperature to compare among member and casual riders. We can see that on average casual rider use the bikes when it is little warmer than the member riders. The number of members is relatively higher than the casual members, which might skew the result - but looks like the range of the temperature that members use bikes is wider than the casual riders. There are some outliers in the casual riders compared to member riders. 

```python
#Create FacetGrid
grid = sns.FacetGrid(df, col = "member_casual")
grid.map(sns.histplot, "avgTemp", bins = 15)
plt.show()
```

Based on FacetGrid, we can see the use of the bikes in different temperature is comparable among member-riders and casual-riders. For both group, the bike ride peaks at 25 degrees and declines when the temperature is less than 0. Since the number of member riders are high, we see the taller bars for member riders than the casual riders.
